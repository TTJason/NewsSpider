```python
from pypinyin import pinyin, lazy_pinyin
import pypinyin
import re
from ltp_model import *
```

```python
LTP_DATA_DIR = '/Users/sivber/Desktop/NLP_DATA/ltp_data'
segmentor = get_segmentor(LTP_DATA_DIR,'lexicon')
postagger = get_postagger(LTP_DATA_DIR)
recognizer = get_recognizer(LTP_DATA_DIR)
parser = get_parser(LTP_DATA_DIR)
labeller = get_labeller(LTP_DATA_DIR)
def get_ltp_model(text):
    return LTPModel(text,
                     segmentor,
                     postagger,
                     recognizer,
                     parser,
                     labeller
                     )
```

```python
rhymes = [{'a','ua','ia'},
          {'e','o','uo','io'},
          {'ve','ie','ue'},
          {'i','v','er'},
          {'u'},
          {'ai','uai'},
          {'ei','ui','uei'},
          {'ao','iao'},
          {'ou','iu','iou'},
          {'an','ian','uan','van'},
          {'en','in','un','uen','vn','n'},
          {'ang','iang','uang'},
          {'eng','ing','ong','iong','ueng'}]
```

```python
titles = open('/Users/sivber/Desktop/NLP_DATA/news/title','r').readlines()
titles = [t.replace('\n','') for t in titles]
```

```python
len(titles)
```

```python
rhyme_collection = [[] for _ in range(len(rhymes))]
word_to_class = {}
for w in titles:
    if len(w.split(' ')) > 0:
        w = w.split(' ')[0]
    py = lazy_pinyin(w,style=pypinyin.FINALS)[-1]
    if py.strip() == '':
        continue
    flag = False
    for rhyme in rhymes:
        if py in rhyme:
            flag = True
            index = rhymes.index(rhyme)
            rhyme_collection[index].append(w)
            word_to_class[w] = index
            continue
    if flag:
        continue
```

```python
expectation = [sum(len(sent) for sent in cls)/len(cls) for cls in rhyme_collection]
```

```python
variance = [(sum((len(sent)-exp)**2 for sent in cls)/len(cls))**(1/2) for cls,exp in zip(rhyme_collection,expectation)]
```

```python
left_rhyme_collection = [[sent for sent in cls if abs(len(sent)-exp) < var] for cls,exp,var in zip(rhyme_collection,expectation,variance)]
```

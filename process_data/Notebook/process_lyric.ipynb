```python
def from_source_to_dev(source_name,dev_name):
    source_file = open(source_name,'r')
    lines = source_file.readlines()[:5]
    dev_file = open(dev_name,'w')
    for line in lines:
        line = line.replace('\n','')
        dev_file.write(line)
        dev_file.write('\n')
    source_file.close()
    dev_file.close()
```

```python
from_source_to_dev('one_word_one_sentence/lyric.source','one_word_one_sentence/dev.source')
from_source_to_dev('one_word_one_sentence/lyric.target','one_word_one_sentence/dev.target')
```

```python
from_source_to_dev('one_word_one_sentence/lyric.source','one_word_one_sentence/test.source')
from_source_to_dev('one_word_one_sentence/lyric.target','one_word_one_sentence/test.target')
```
